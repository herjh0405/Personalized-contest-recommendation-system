# Simple Recommender
유저는 처음 가입 시 선호 분야와 신분을 기입하고 시작함

기본적으로 접수 중인 대회들을 추천해줘야 하고, 후 순위로 지나간 대회를 추천해줘야 함
* 조회수/접수기간, 상금을 조합하여서 순위를 정해야 함
* 크롤링 시 접수중인지 아닌지 나눌 필요성이 있고, 상금을 제대로 뽑아내어야하며, 조회기간도 잘못 된 경우가 꽤 발생

```python
import pandas as pd
import re
```

```python
df_last = pd.read_csv('./data/최종.csv')
```

```python
df_last.columns
```

```python
df_last['1등시상금'] = df_last['1등시상금'].apply(lambda x : x.replace('만원', '0000'))
```

```python
df_last['1등시상금'] = df_last['1등시상금'].apply(lambda x : re.sub('[^ 0-9]', '', x))
```

```python
df_last['1등시상금'].unique()[:30]
```

```python
df_last['1등시상금'].unique()
```

```python
re.sub('[^ 0-9]', df_last['1등시상금'])
```

```python
df_last['접수종료'] = df_last['접수기간'].apply(lambda x: pd.to_datetime(x.replace(' ','').split('~')[1]))
df_last['접수시작'] = df_last['접수기간'].apply(lambda x: pd.to_datetime(x.replace(' ','').split('~')[0]))
```

```python
df_last['접수'] = (df_last['접수종료'] - df_last['접수시작']).dt.days
```

```python
df_last = df_last.loc[df_last['접수']>0]
```

```python
df_last['inter'] = (df_last['조회수']/df_last['접수']).astype('int')
```

```python
df_last.sort_values(['inter'], ascending = False)
```
