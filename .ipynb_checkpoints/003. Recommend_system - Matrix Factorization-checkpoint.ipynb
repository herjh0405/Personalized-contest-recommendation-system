```python
user_id = [rd.randint(1, 10000) for i in range(1, 100001)]
gong_rank = [rd.randint(1, 10) for i in range(1, 100001)]
gong_id = [rd.randint(1, 14450) for i in range(1, 100001)]

user_id.sort()

user_info = pd.DataFrame({'user_id' : user_id,
             'gong_id' : gong_id})
user_info = user_info[['user_id','gong_id']].drop_duplicates()

user_info['gong_rank'] = gong_rank = [rd.randint(1, 10) for i in range(1, len(user_info)+1)]
```

## 데이터 불러오기

```python
user_info = pd.read_csv('./data/user_info.csv')
gong_info= pd.read_csv('./data/최종.csv')
```

## 분류에 따른 아이템별 추천

### 함수 사용법
* gongmo_recommend(juchae, money, title) : 
   고른 아이템과 비슷한 공모전들을 추천해줌
    * juchae : 주최기관, money : 상금규모, title : 공모전명

```python
def gongmo_recommend(juchae, money, title):
    # 원하는 정보만 뽑음
    gong_temp = gong_info[gong_info[juchae]==1].reset_index(drop = True)
    gong_want = gong_temp[gong_temp[money]==1].reset_index(drop = True)
    
    # rank 기준
    user_data = user_info.copy()
    gong_data = gong_want[['gong_id', '공모전명']].copy()
    
    # SVD 적용하기 좋은 모습으로 변경
    user_gong_data = pd.merge(user_data, gong_data, on = 'gong_id')
    user_gong_rating = user_gong_data.pivot_table('gong_rank',
                    index = 'user_id', columns = '공모전명').fillna(0)
    user_user_rating = user_gong_rating.values.T
    
    # 시그마에서 상위 12개 element만 추출
    SVD = TruncatedSVD(n_components = 12)
    matrix = SVD.fit_transform(user_user_rating)
    
    # 상관계수 도출
    corr = np.corrcoef(matrix)
    
    # 시각화
    plt.figure(figsize=(16, 10))
    sns.heatmap(corr)
    plt.show()
    
    gong_title = user_gong_rating.columns
    gong_title_list = list(gong_title)
    coffey_hands = gong_title_list.index(title)
    
    # 비슷할 확률이 80% 이상인 공모전들만 반환
    corr_coffey_hands  = corr[coffey_hands]
    recommend_list = list(gong_title[(corr_coffey_hands >= 0.8)])
    if title in recommend_list :
        recommend_list.remove(title)
        
    return recommend_list
```

```python
gongmo_recommend('정부/공공기관', '5000', '제4회 ICT 스마트디바이스전국 공모전')
```

사용자 - 공모전별 랭킹 pivot table 생성

```python
df_user_gong_ratings = user_info.pivot(index = 'user_id', columns = 'gong_id',
                                      values = 'gong_rank').fillna(0)
```

```python
df_user_gong_ratings.head()
```

위의 데이터를 살짝 변형 시켜 사용하겠다.
1. pivot table을 matrix로 변환
2. np.mean(axis = 1)을 통해 각 사용자들이 매기는 평점 평균을 구함
3. 1에서 구한 값과 2에서 구한 값을 빼서 사용자 - 평균 데이터 값을 변경

```python
# matrix는 pivot_table 값을 numpy matrix로 만든 것 
matrix = df_user_gong_ratings.values

# user_ratings_mean은 사용자의 평균 평점 
user_ratings_mean = np.mean(matrix, axis = 1)

# R_user_mean : 사용자-영화에 대해 사용자 평균 평점을 뺀 것.
# 모든 열에 다 빼준다
matrix_user_mean = matrix - user_ratings_mean.reshape(-1, 1)
```

여기까지 진행하면 초기에 만들었던 사용자-공모전 pivot table 값이
matrix_user_mean으로 변경되었음.

1. 사용자 9999명이 14427개의 공모전에 대해 평점을 매긴 값이 존재
2. 사용자 9999명의 각각 평균 평점을 구해서
3. 사용자 9999명이 14427개의 영화에 대한 평점을 변경 -> 1에서 구한 값 - 2에서 구한 값

이제 SVD를 이용해 Matrix Factorization을 진행.

앞서서는 scikit learn 대신 이번에는 scipy를 이용해 Truncated SVD를 구해봄.

이 둘의 차이점은 scikit learn에서 제공해주는 TruncatedSVD는 U, Sigma, Vt 반환 값을 제공하지 않지만 Scipy를 이용하면 이 반환값들을 제공받을 수 있음.

Scipy에서 제공해주는 Truncate SVD는 scipy.sparse.linalg.svds를 사용.

```python
# scipy에서 제공해주는 svd.  
# U 행렬, sigma 행렬, V 전치 행렬을 반환.

U, sigma, Vt = svds(matrix_user_mean, k = 12)
```

```python
print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

현재 이 Sigma 행렬은 0이 아닌 값만 1차원 행렬로 표현된 상태이다.
즉, 0이 포함 된 대칭행렬로 변환시키기 위해 numpy의 diag 사용

```python
sigma = np.diag(sigma)
```

```python
sigma.shape
```

이렇게 대칭 행렬로 변환이 되었음.

현재까지 상황을 정리하면 아래와 같다.

1. 원래 사용자 - 공모전 평점 행렬이 있었음
2. 이를 user의 평균 점수를 빼서 matrix_user_mean 이라는 행렬로 만듬
3. 2번의 값을 SVD를 적용해 U, Sigma, Vt 행렬을 구했음
4. Sigma 행렬은 현재 0이 포함이 되지 않은 값으로만 구성되어 있음. 이를 대칭행렬로 변환

이제 여기서 matrix_user_mean을 SVD를 적용해 분해를 한 상태이다.
다시 원본 행렬로 복구시키기 위해 아래를 시행한다.

1. U, Sigma, Vt의 내적을 수행 즉, np.dot(np.dot(U, sigma), Vt)를 수행해준다.
2. 아까 사용자 평균을 빼주었으니 여기서는 더해준다.

```python
# U, Sigma, Vt의 내적을 수행하면, 다시 원본 행렬로 복원이 된다. 
# 거기에 + 사용자 평균 rating을 적용한다. 
svd_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
```

```python
df_svd_preds = pd.DataFrame(svd_user_predicted_ratings, columns = df_user_gong_ratings.columns)
df_svd_preds.head()
```

## 개인별 맞춤형 공모전 추천 시스템 함수 생성

1. 인자로 사용자 아이디(user_id), 추천 갯수(num_recommendations)를 받음 
2. 사용자 아이디에 SVD로 나온 결과의 영화 평점이 가장 높은 데이터 순으로 정렬
3. 사용자가 본 데이터를 제외
4. 사용자가 안 본 영화에서 평점이 높은 것을 추천

```python
# num_recommendations : 추천해주는 갯수
def recommend_movies(user_id, num_recommendations=5):
    
    #현재는 index로 적용이 되어있으므로 user_id - 1을 해줌.
    user_row_number = user_id - 1 
    
    # 최종적으로 만든 pred_df에서 사용자 index에 따라 영화 데이터 정렬 -> 영화 평점이 높은 순으로 정렬 됨
    sorted_user_predictions = df_svd_preds.iloc[user_row_number].sort_values(ascending=False)
    
    # 원본 평점 데이터에서 user id에 해당하는 데이터를 뽑아낸다. 
    user_data = user_info[user_info.user_id == user_id]
    
    # 위에서 뽑은 user_data와 원본 영화 데이터를 합친다. 
    user_history = user_data.merge(gong_info, on = 'gong_id').sort_values(['gong_rank'], ascending=False)
    
    # 원본 영화 데이터에서 사용자가 본 영화 데이터를 제외한 데이터를 추출
    recommendations = gong_info[~gong_info['gong_id'].isin(user_history['gong_id'])]
    
    # 사용자의 영화 평점이 높은 순으로 정렬된 데이터와 위 recommendations을 합친다. 
    recommendations = recommendations.merge( pd.DataFrame(sorted_user_predictions).reset_index(), on = 'gong_id')
    
    # 컬럼 이름 바꾸고 정렬해서 return
    recommendations = recommendations.rename(columns = {user_row_number: 'Predictions'}).sort_values('Predictions', ascending = False).iloc[:num_recommendations, :]
                      

    return user_history, recommendations
```

```python
already_rated, predictions = recommend_movies(3588, 10)
```
